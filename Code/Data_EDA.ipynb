```python
data.isnull().sum()

# To check for missing data in each feature   
# isnull returns the entire dataframe input itself as it is with True in places where null (missing values) exist and True otherwise
```

```python
# Categorical Data Conversion

for i in data:
    print(data[i].value_counts())   # Shows list of unqiue data for each feature

for i in data:
    print(data[i].value_counts())   # Shows the count of each data under a feature
```

```python
cat_df = data.select_dtypes(include=['object'])  # Extracts only Categorical data from the Dataset, This method only needs 1 parameter, Include(or Exclude)

for i in cat_df:
    print(cat_df[i].value_counts()) 
```
